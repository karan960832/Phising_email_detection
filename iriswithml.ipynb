```python
import pandas as pd 
import numpy as np
```

```python
from sklearn import metrics
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn import svm
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
iris = pd.read_csv(r"C:\Users\ASUS\Downloads\IRIS.csv")
```

```python
iris.head
```

```python
iris.describe()
```

```python
train,test=train_test_split(iris,test_size=0.25)
train
```

```python
test
```

```python
train_x=train[['sepal_length','sepal_width','petal_length','petal_width']]
```

```python
train_y=train['species']
```

```python
train_x
```

```python
train_y
```

```python
test_x=test[['sepal_length','sepal_width','petal_length','petal_width']]
```

```python
test_y=test['species']
```

```python
test_x
```

```python
test_y
```

```python
#model 1
model=svm.SVC()
model.fit(train_x, train_y)
predictions=model.predict(test_x)
predictions
```

```python
metrics.accuracy_score(test_y, predictions)
```

```python
#decision tree classifier
model2=DecisionTreeClassifier()
model2.fit(train_x, train_y)
predictions2=model2.predict(test_x)
predictions2
```

```python
metrics.accuracy_score(test_y, predictions2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Calculate confusion matrix
cm = confusion_matrix(test_y, predictions)

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_y, predictions))
```

```python
import joblib
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
# Train a dummy Iris model
iris = load_iris()
clf = RandomForestClassifier()
clf.fit(iris.data, iris.target)

# Save the model and classes
joblib.dump(clf, 'species_model.pkl')
joblib.dump(iris.target_names, 'species_classes.pkl')
print("Test models created successfully!")
```
